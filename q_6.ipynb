```python
import numpy as np
import matplotlib.pyplot as plt

# define the function
def f(x):
    return 10
    
xmin = -10
xmax = 10
n = 256
dx = (xmax - xmin)/(n-1)

# array of x values
xarr = np.linspace(xmin, xmax, n)

# array for f(x) values
fxarr = [f(i) for i in xarr]

# DFT of fxarr
dft_arr = np.fft.fft(fxarr, norm="ortho")

# array of k values
karr = 2 * np.pi * np.fft.fftfreq(n, dx)

# Calculate  F(K)
fkarr =  np.zeros(len(dft_arr), dtype=np.complex128)     # array of F(K)

for i in range(len(karr)):
    fkarr[i]=dx*np.sqrt(n/(2*np.pi))*np.exp(-(1j)*karr[i]*xmin)*(dft_arr[i])


# Plot the  function and its Fourier transform
plt.figure(figsize=(12, 6))

# Plot f(x)
plt.subplot(1, 2, 1)
plt.plot(xarr, fxarr, label = 'f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title("Plot of f(x)")
plt.legend()
plt.grid(True, alpha = 0.5)

# Plot F(k)
plt.subplot(1, 2, 2)
plt.plot(karr, fkarr.real, color = 'r', label = "FT of f(x)")
plt.xlabel('k')
plt.ylabel('F(k)')
plt.title(" Fourier transform of f(x)")
plt.legend()
plt.grid(True, alpha = 0.5)

plt.show()
```
